```python
import torch
from transformers import AdamW, AutoTokenizer, AutoModelForSequenceClassification

# Same as before
checkpoint = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(checkpoint, token=os.environ.get('HF_TOKEN'))
model = AutoModelForSequenceClassification.from_pretrained(checkpoint, token=os.environ.get('HF_TOKEN'))
sequences = [
    "I've been waiting for a HuggingFace course my whole life.",
    "This course is amazing!",
]
batch = tokenizer(sequences, padding=True, truncation=True, return_tensors="pt")

# This is new
batch["labels"] = torch.tensor([1, 1])

optimizer = AdamW(model.parameters())
loss = model(**batch).loss
loss.backward()
optimizer.step()
```

```python
from datasets import load_dataset

raw_datasets = load_dataset("AI-Secure/adv_glue", "adv_mnli")
```

```python
raw_datasets
```

```python
raw_datasets['validation'][0]
```

```python
def tokenize_function(example):
    return tokenizer(example["premise"], example["hypothesis"], truncation=True)
```

```python
from transformers import AutoTokenizer

checkpoint = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(checkpoint, token=os.environ.get('HF_TOKEN'))
```

```python
tokenized_datasets = raw_datasets.map(tokenize_function, batched=True)
tokenized_datasets
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
samples = tokenized_datasets["validation"][:8]
samples = {k: v for k, v in samples.items() if k not in ["idx", "premise", "hypothesis"]} # remove unneccesary columns
[len(x) for x in samples["input_ids"]]
```

```python
batch = data_collator(samples)
{k: v.shape for k, v in batch.items()}
```
